```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dt = pd.read_csv(r"/content/House_hold_data.csv.csv")

dt
```

```python
# scatter plot with year against own
plt.scatter(dt['year'], dt['own'])
plt.title('scatter plot')
plt.xlabel('year')
plt.ylabel('own')
plt.show()
```

```python
# line chart with year and own
plt.plot(dt['year'])
plt.plot(dt['own'])

plt.title('Scatter chart')
plt.xlabel('year')
plt.ylabel('own')
plt.show()
```

```python
# bar chart
plt.bar(dt['year'], dt['own'])
plt.title('bar chart')
plt.xlabel('year')
plt.ylabel('own')
plt.show()
```

```python
# histogram
plt.hist(dt['income'])
plt.title('histogram')
plt.xlabel('income')
plt.ylabel('frequency')
plt.show()
```
